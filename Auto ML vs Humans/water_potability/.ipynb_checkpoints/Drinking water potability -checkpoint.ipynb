```python
X = df.drop('Potability', axis= 1)
y = df['Potability']
```

Lets just do some feature significance viz because why not

```python
from sklearn.ensemble import ExtraTreesClassifier 
```

```python
etc = ExtraTreesClassifier()
etc.fit(X,y)
```

```python
features = pd.Series(etc.feature_importances_,index = X.columns)
features.sort_values().plot(kind = 'barh')
```

### Scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
scale = scale.fit_transform(X)
scale 
```

```python
scaled_df = pd.DataFrame(scale,columns = X.columns)
scaled_df
```

### Modelling 
Lets do some modelling. I will try a few different models nothing too fancy 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_df, y, test_size = 0.25)
```

```python
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report, accuracy_score
```

```python
def metric_scores(pred):
    print(f"ROC AUC: {roc_auc_score(y_test, pred)*100:.2f}")
    print(f"ACC SCORE: {accuracy_score(y_test, pred)*100:.2f}")
    print(f"CONFUSION MATRIX: {confusion_matrix(y_test, pred)}")
    print(f"CLASSIFICATION REPORT: {classification_report(y_test, pred)}")
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression()
lgr.fit(X_train, y_train)
lgr_predict = lgr.predict(X_test)
```

```python
metric_scores(lgr_predict)
```

### SVM-SVC

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(X_train, y_train)
svc_pred = svc.predict(X_test)
```

```python
metric_scores(svc_pred)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators' : [100,200,500],
              'max_depth' : [3 , 4, 5, 6],
              'max_features' : ['auto','sqrt','log2'],
              'criterion' : ['gini', 'entropy']
             }
```

```python
param_model = GridSearchCV(RandomForestClassifier(), param_grid = param_grid, scoring = 'roc_auc', verbose = 1)
```

```python
param_model.fit(X_train, y_train)
```

```python
param_model.best_params_
```

```python
rfc = RandomForestClassifier(criterion = 'gini', max_depth = 6, max_features = 'log2', n_estimators = 200, verbose = 2)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
metric_scores(rfc_pred)
```

This is one beefy chunky bad boy so beware when you are fitting the model

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init(max_mem_size='8G')
```

```python
df = h2o.import_file('/home/thayyeb/Desktop/Project/water_potability/water_potability.csv')
df['Potability'] = df['Potability'].asfactor()
```

```python
df_train, df_test= df.split_frame(ratios= [.8])
```

```python
y = 'Potability'
x = df.columns
x.remove(y)
x
```

```python
aml = H2OAutoML(max_models=10, seed = 10, verbosity = 'info', nfolds= 2)
```

```python
aml.train(x=x, y = y, training_frame=df_train)
```

```python
lb = aml.leaderboard
lb
```

```python
model_ids = list(aml.leaderboard['model_id'].as_data_frame().iloc[:,0])
model_ids
```

```python
aml.leader.model_performance(df_test)
```

```python
aml.leader
```

```python
aml_pred = aml.leader.predict(df_test)
aml_pred
```

## Eval ML
This is our usual challenger maybe sometime in the future he will defect camps. Terminator flashback

```python
import evalml
from evalml import AutoMLSearch
```

```python
df = pd.read_csv('/home/thayyeb/Desktop/Project/water_potability/water_potability.csv')
X = df.drop('Potability', axis = 1)
y = df['Potability']
```

```python
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X, y, test_size = 0.3, problem_type = 'binary')
```

```python
aml = AutoMLSearch(X_train = X_train, y_train = y_train, problem_type = 'binary', objective = 'AUC')
aml.search()
```

```python
aml.rankings
```

```python
bestpipeline = aml.best_pipeline
```

```python
aml.describe_pipeline(aml.rankings.iloc[0]['id'])
```

```python
bestpipeline.score(X_test, y_test, objectives = ['AUC'])
```

Well considering how all models had almost the same levels of performance.
